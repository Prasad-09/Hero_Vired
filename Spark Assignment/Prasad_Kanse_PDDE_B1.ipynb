```python
#36.Write 2–3 lines explaining what a broadcast join is and why it is useful when one table is relatively small.
```

```python
#A broadcast join sends the smaller DataFrame to all worker nodes, so it can be joined locally with partitions of the larger DataFrame.
#This avoids expensive data shuffling across the cluster, which is usually the biggest performance bottleneck.
#It’s especially useful when one table is small enough to fit in memory, making joins significantly faster.
```

```python
#Find the top 3 categories in each state.
```

```python
from pyspark.sql.functions import count, desc, row_number
from pyspark.sql.window import Window

window_spec = Window.partitionBy("state").orderBy(desc("category_count"))

top_categories_per_state = final_df.groupBy("state", "category") \
    .agg(count("*").alias("category_count")) \
    .withColumn("rank", row_number().over(window_spec)) \
    .filter(col("rank") <= 3)

top_categories_per_state.show()
```
